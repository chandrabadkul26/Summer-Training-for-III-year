```python
from sklearn.preprocessing import LabelEncoder,StandardScaler

enc = LabelEncoder()

df['CUST_ID'] = enc.fit_transform(df['CUST_ID'])

```

```python
df
```

```python

sc = StandardScaler()
df.iloc[:,2:] = sc.fit_transform(df.iloc[:,2:])
```

```python
df
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
from sklearn.cluster import KMeans
wcss = []
clus = [i for i in range(1,11)]

for i in clus:
    km = KMeans(n_clusters = i,init = 'k-means++')
    km.fit(df.values)
    wcss.append(km.inertia_)
```

```python
plt.plot(clus,wcss)
plt.xlabel("Clusters")
plt.ylabel("WCSS")
```
